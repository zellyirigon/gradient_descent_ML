**Gradient Descent** is an optimisation algorithm used in machine learning to minimise the cost function, which measures how well a model's predictions match the actual data. It works by iteratively adjusting the model's weights to reduce the cost.

- **Initialise weights:** Start with random weights.

- **Calculate gradient:** Compute the gradient of the cost function with respect to the weights. This gradient indicates the direction and rate of the steepest ascent in the cost function.

- **Update weights:** Adjust the weights in the opposite direction of the gradient to reduce the cost. This step is controlled by the learning rate, a small positive number.

- **Repeat:** Continue this process until the cost function reaches a minimum or sufficiently low value.

**The goal is to find the optimal weights that minimise the cost function, leading to the best model performance.**

### Example
Find the local minimum of the function y =(x + 5)² starting from the point x =3
It's easy to find the answer just by looking at the chart. y = (x + 5)² reaches its minimum value when x = -5(i.e., when x = -5, y =0). 
Therefore, x = -5 is the local and global minimum of the function

### Applying the Gradient Descent 

```python
# The algorithm starts with parameter x = 3
cur_x = 3

# Learning rate - Regulate the size of the step
rate = 0.03

# Defining when to stop the algorithm
# Stopping the loop when the difference between the x values in 2 consecutive iterations is less than 0,000001
# or when the iteration number exceeds 10k.
precision = 0.000001

# Initialising the count of the previous step
previous_step_size = 1

# Maximum number of iterations
max_iters = 10000

# Iteration count
iters = 0

# Gradient of the function
df = lambda x: 2 * (x+5)
```

```python
while previous_step_size > precision and iters < max_iters:
    # Saving the current value of the x in pred_x
    pred_x = cur_x
    
    # Applying the gradient descent - It defines the new x value
    cur_x = cur_x - rate * df(pred_x)
    
    # Changing the x value -> (cur_x - pred_x) means the comparison between observed values(output data) and the value predicted by the model. The best
    #scenario is the result equals zero, which means I'm finding the ideal value of x.
    previous_step_size = abs(cur_x - pred_x)
    
    # Incrementing the iteration number
    iters +=1
    
    # Printing the iterations
    print('Iteration' , iters, '\nX Value = ', cur_x)
print('\n The function local minimum occurs in: ', cur_x)
```
